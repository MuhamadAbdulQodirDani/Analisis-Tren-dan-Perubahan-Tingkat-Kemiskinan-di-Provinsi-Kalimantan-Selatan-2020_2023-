#Analisis Tren dan Perubahan Tingkat Kemiskinan di Provinsi Kalimantan Selatan (2020-2023)

##Latar Belakang
Kemiskinan merupakan salah satu masalh utama yang dihadapi oleh provinsi Kalimantan Selatan. Pemahaman yang mendalam tentang perubahan tingkat kemiskinan dari tahun ke tahun sangat penting bagi pemerintah daerah untuk merancang kebijakan yang tepat sasaran. Analisis ini bertujuan untuk mengeksplorasi tren dan perubahan tingkat kemiskinan di berbagai kabupaten di Kalimantan Selatan dari tahun 2020 hingga 2023.

##Tujuan Analisis
1. Mengidentifikasi tren tingkat kemiskinan di Kalimantan Selatan dari tahun 2020 hingga 2023.
2. Menentukan wilayah dengan peningkatan dan penurunan tertinggi dalam tingkat kemiskinan.
3. Memberikan rekomendasi inovatif berdasarkan hasil analisis data untuk membantu pemerintah dan masyarakat dalam mengatasi masalah kemiskinan.

##Analisis Data dan Interpretasi

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Memuat Dataset
data = pd.read_csv('/content/tingkat kemiskinan 2020-2023.csv')
data
```

```python
# Menghapus data kosong
df = data.dropna()
df
```

```python
# Satistik Deskriptif
df.describe()
```

```python
# Melebur dataframe untuk memudahkan plotting
df_melted = df.melt(id_vars=['Kabupaten'], var_name='Tahun', value_name='Tingkat Kemiskinan')
df_melted.head(), df_melted.tail()
```

```python
# Distribusi Tingkat Kemiskinan pada Tahun 2023
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='2023', bins=10, kde=True)
plt.title('Distribusi Tingkat Kemiskinan pada Tahun 2023')
plt.xlabel('Tingkat Kemiskinan (%)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Tren Tingkat Kemiskinan dari 2020 - 2023
plt.figure(figsize=(10,6))
sns.lineplot(data=df_melted, x='Tahun', y='Tingkat Kemiskinan', hue='Kabupaten', marker='o')
plt.title('Tren Tingkat Kemiskinan dari 2020 - 2023')
plt.xlabel('Tahun')
plt.ylabel('Tingkat Kemiskinan (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Heatmap Korelasi antar Tahun
correlation = df.drop(columns='Kabupaten').corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Korelasi Tingkat Kemiskinan antar Tahun')
plt.show()
```

```python
# Analisis Perubahan Tingkat Kemiskinan
df['Change_2020_2023'] = df['2023'] - df['2020']
print("\nPerubahan Tingkat Kemiskinan dari 2020 hingga 2023:\n", df[['Kabupaten', 'Change_2020_2023']])
```

```python
# Wilayah dengan Peningkatan dan Penurunan Tertinggi dalam tingkat kemiskinan
max_increase = df.loc[df['Change_2020_2023'].idxmax()]
max_decrease = df.loc[df['Change_2020_2023'].idxmin()]

print("\nWilayah dengan peningkatan tertinggi dalam tingkat kemiskinan dari 2020 hingga 2023:\n", max_increase)
print("\nWilayah dengan penurunan tertinggi dalam tingkat kemiskinan dari 2020 hingga 2023:\n", max_decrease)
```

###Interpretasi Hasil Analisis
1. Statistik Deskriptif: Menunjukkan distribusi data tingkat kemiskinan dari tahun 2020 hingga 2023, termasuk rata-rata, standar deviasi, nilai minimum, dan maksimum.
2. Distribusi Tingkat Kemiskinan di Tahun 2023: Histogram menunjukkan bahwa sebagian besar wilayah memiliki tingkat kemiskinan yang relatif rendah, dengan beberapa wilayah mengalami tingkat kemiskinan yang lebih tinggi.
3. Tren Tingkat Kemiskinan dari 2020 hingga 2023: Plot garis menunjukkan perubahan tingkat kemiskinan di berbagai wilayah dari tahun 2020 hingga 2023. Beberapa wilayah menunjukkan tren penurunan, sementara yang lain menunjukkan peningkatan.
4. Korelasi Antar Tahun: Heatmap menunjukkan bahwa ada korelasi yang kuat antara tingkat kemiskinan dari tahun ke tahun, yang menunjukkan bahwa wilayah dengan tingkat kemiskinan tinggi di satu tahun cenderung memiliki tingkat kemiskinan tinggi di tahun berikutnya.
5. Perubahan Tingat Kemiskinan dari Tahun 2020 hingga 2023: Menunjukkan wilayah dengan peningkatan dan penurunan tertinggi dalam tingkat kemiskinan. Wilayah dengan peningkatan tertinggi memerlukan perhatian khusus untuk mengetahui penyebabnya, sementara wilayah dengan penurunan tertinggi dapat menjadi contoh praktik terbaik.
